```python
# Load PCs

from pathlib import Path
import pandas as pd

PROJECT = Path.cwd()
PROCESSED = PROJECT / "data" / "processed"

pcs = pd.read_csv(PROCESSED / "pca_20pcs.csv", index_col=0)
print("pcs:", pcs.shape)
```

```python
# Fit k-means with k=2 (fixed seed)
from sklearn.cluster import KMeans

k = 2
km = KMeans(n_clusters=k, n_init=25, random_state=1)
labels = km.fit_predict(pcs)

subtypes_k2 = pd.DataFrame({"subtype": labels}, index=pcs.index)
subtypes_k2["subtype"] = subtypes_k2["subtype"].map(lambda x: f"C{x+1}")

print(subtypes_k2["subtype"].value_counts())

```

```python
# Save labels (
subtypes_k2.to_csv(PROCESSED / "subtype_labels_k2.csv")
print("Saved: subtype_labels_k2.csv")

subtypes_k2.to_csv(PROCESSED / "subtype_labels.csv")
print("Overwrote: subtype_labels.csv (now k=2)")
```
