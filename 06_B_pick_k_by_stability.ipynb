# Load PCs
from pathlib import Path
import pandas as pd

PROJECT = Path.cwd()
PROCESSED = PROJECT / "data" / "processed"
pcs = pd.read_csv(PROCESSED / "pca_20pcs.csv", index_col=0)



```python
# Trying k = 2 to 6 and summarise stability (ARI) since the previous k= 3 was not that stable.

from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import numpy as np

def stability_for_k(pcs, k, seeds=range(1, 21)):
    ref = KMeans(n_clusters=k, n_init=25, random_state=1).fit_predict(pcs)
    aris = []
    for seed in seeds:
        labels = KMeans(n_clusters=k, n_init=25, random_state=seed).fit_predict(pcs)
        aris.append(adjusted_rand_score(ref, labels))
    return float(np.min(aris)), float(np.median(aris)), float(np.max(aris))

results = []
for k in range(2, 7):
    mn, med, mx = stability_for_k(pcs, k)
    results.append((k, mn, med, mx))

stability = pd.DataFrame(results, columns=["k", "min_ARI", "median_ARI", "max_ARI"])
stability
```

```python
best = stability.sort_values("min_ARI", ascending=False).head(1)
best
```
